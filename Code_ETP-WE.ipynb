```python
def cpuUtilization(p):
    rcpu=0
    for i in p:
        #print('from cpuutilization')
        x=(i.status * i.allo_cpu )/i.MIPS
        rcpu = rcpu + x 
    print('cpuUtilization',rcpu)
    return(rcpu)
```

```python
def memoryUtilization(p):
    rmem=0
    for i in p:
        #print('from memoryutilization')
        x=(i.status * i.allo_mem)/i.MEMORY
        rmem = rmem + x 
    print('memoryUtilization',rmem)
    return(rmem)
```

```python
def resourceUtilization(p):
    #print('resource utilization called')
    c=0
    Rdc=0
    for i in p:
        if(i.status==1):
            
            c=c+1
    if c==0:
        c=1#print('RU',c)
    f=cpuUtilization(p) + memoryUtilization(p)
    g=2*c
    Rdc= (f)/(g)
    print('RESOURCE UTILIZATION',Rdc)
    print('Number of active servers',c)
    return(Rdc)
    
```

```python
def VMmigration(p,src,vm,mig):
    if (vm.status=='not allocated'):
        h=pm=0
        for i in range(pl):
            if p[i].id==src:
                print('CPU and memmory added to source PM after migration')
                p[i].mips=p[i].mips-vm.mips
                p[i].memory= p[i].memory-vm.memory
                
        if src> pl/2:
            for h in range(src,0,-1):
                if p[h].status==1:
                    mapping(p[h], vm)
                if (vm.status=='allocated'):
                    pm=p[h].id
                    ############# Calculate VM migration cost ################
                    mig=abs(src-p[h].id)*100*vm.mips*vm.memory + mig
                    print('From PM',src,'to destination Active PM',p[h].id,'total VM migration cost', mig)   
                    break
                if (vm.status==' not allocated'):
                    h=0
                    for h in range(src,0,-1):
                    ####### waking up PM having largest capacity for VM migration ############
                        if p[h].status==0:
                            mapping(p[h], vm)
                        if (vm.status=='allocated'):
                            pm=p[h].id
                            mig=abs(src-p[h].id)*100*vm.mips*vm.memory+ 4260 + mig
                            print('From PM',src,'to destination New waken up PM',p[h].id,'total VM migration cost', mig)
                            break
        else:
            h=src
            for h in range(pl):
                    ###########try to use already active PM#########
                if p[h].status==1:
                    mapping(p[h], vm)
                if (vm.status=='allocated'):
                    pm=p[h].id
                    ############# Calculate VM migration cost ################
                    mig=abs(src-p[h].id)*100*vm.mips*vm.memory + mig
                    print('From PM',src,'to destination Active PM',p[h].id,'total VM migration cost', mig)   
                    break
                if (vm.status==' not allocated'):
                    h=src
                    for h in range(pl):
                    ####### waking up PM having largest capacity for VM migration ############
                        if p[pl-h].status==0:
                            mapping(p[pl-h], vm)
                        if (vm.status=='allocated'):
                            pm=p[pl-h].id
                            mig=abs(src-p[h].id)*100*vm.mips*vm.memory+ 4260 + mig
                            print('From PM',src,'to destination New waken up PM',p[h].id,'total VM migration cost', mig)
                            break
                    
                       
                        
    
    
    
    #print('total VM migration cost', mig)                        
    return p,pm,mig                                                    
                            
                            
        
        
        
        
        
                
    
```

```python
### VM vulnerability score
def riskR1(v,vm): 
    
    for i in range(vl):
        if vm.id==v[i].id:
            r= choice([a for a in range(0,10)]) 
            r1=r/10
            v[i].score=r1 
            break
        
    #print('r1',r1)   
    return r1       
    
    
```

```python
##### Hypervisor vulnerability score
def riskR2(p,v,vm):
    arr=[]
    r2=0
    d=vm.allocatedTo
    #print('Risk 2 for VM',vm.id,'on PM',d)
    flag=0
    for i in range(pl):
        if d==p[i].id:
            vx = [virtualMachines() for j in range(len(p[i].alloVM))]  
            r= choice([a for a in range(0,10)]) 
            s=r/10 
            
            for k in range(len(p[i].alloVM)):
                ik=int(p[i].alloVM[k])
                v[ik].score=riskR1(v,v[ik])
                vx[k]=v[ik]
                arr.append(v[ik].score)
            #print('RISK 2 arr',arr)
            if len(arr)>0:
                m=max(arr)
            else:
                m=0.5
            r2=s*(0+ m)
            p[i].hscore=r2
            flag=1
            break
        if flag==1:
            break
            
            
           
    #print('r2',r2)
    return r2
            
```

```python
##### Co-residency #####
def riskR3(p,v,vm):
    pe=1
    d=vm.allocatedTo
    for j in range(pl):
        if p[j].id==d:
            for k in range(len(p[j].alloVM)):
                h=int((p[j]).alloVM[k])
                #print(' RISK 3 h=p[j].alloVM[k]', h)
                e=riskR1(v,v[h])
                pe=pe*(1-e)
    r3=1-pe
    #print('r3',r3)
    return r3
                
            
    
    
```

```python
####Network vulnerability  #####
def riskR4(p,v,vm,user):
    pe=1
    for i in range(vl):
        if v[i].user== user:
            v[i].score=riskR1(v,v[i])
            pe=pe*(1-v[i].score)
    r4=1-pe
    #print('r4',r4)
    return r4
            
    
    
```

```python
def totrisk(vm,v,p,user):
    r=1-((1-riskR4(p,v,vm,user))*(1-riskR3(p,v,vm))*(1-riskR2(p,v,vm))*(1-riskR1(v,vm)))
    return r
```

```python
def riskmatrix_score(v,p):
    for i in range(vl):
        v[i].r1=riskR1(v,v[i])
        v[i].r2=riskR2(p,v,v[i])
        v[i].r3=riskR3(p,v,v[i])
        v[i].r4=riskR4(p,v,v[i],v[i].user)
        
        v[i].risk=totrisk(v[i],v,p,v[i].user)
        print('Risk matrix score for VM',v[i].id,'is',v[i].risk)
        print('VM vulnerability',v[i].r1,'Hypervisor vulnerability',v[i].r2,'Network vulnerability', v[i].r4, 'Coresidency',v[i].r3) 
        
```
